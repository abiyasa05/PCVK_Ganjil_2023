6. Lakukan deteksi bagian wajah gambar “mjordan.jpg”, kemudian tindih bagian wajah tersebut
dengan gambar “smile.png”. Berikut contoh keluarannya.

```python
# Load the image with faces
image_path = '/content/drive/MyDrive/PCVK/Images/Face Detection/mjordan.jpg'
img = cv.imread(image_path)
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Load the pre-trained face cascade classifier
face_cascade = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Check if the classifier is loaded successfully
if face_cascade.empty():
    print("Error loading face cascade classifier.")
    exit()

# Convert the image to grayscale for face detection
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Detect faces in the image
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5, minSize=(30, 30))

# Load the smile image with an alpha channel
smile_path = '/content/drive/MyDrive/PCVK/Images/Face Detection/smile.png'
smile_img = cv.imread(smile_path, cv.IMREAD_UNCHANGED)

# Check if the smile image is loaded successfully
if smile_img is None:
    print("Error loading smile image.")
    exit()

# Overlay smile on detected faces
for (x, y, w, h) in faces:
    # Resize smile image to fit the detected face
    resized_smile = cv.resize(smile_img, (w, h))

    # Extract alpha channel from the smile image (if it exists)
    if resized_smile.shape[2] == 4:
        alpha_smile = resized_smile[:, :, 3] / 255.0

        # Extract the region of interest (ROI) from the original image
        roi = img[y:y+h, x:x+w, 0:3]

        # Resize alpha channel to match the ROI size
        alpha_smile = cv.resize(alpha_smile, (w, h))

        # Blend the smile and the ROI using alpha blending
        for c in range(0, 3):
            roi[:, :, c] = roi[:, :, c] * (1 - alpha_smile) + resized_smile[:, :, c] * alpha_smile

# Display the result
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

7. Lakukan deteksi wajah dengan menggunakan bounding box berwarna hitam, deteksi mata
menggunakan bounding box berwarna putih, dan deteksi mata kanan menggunakan bounding box
berwarna merah. Gunakan gambar “people.jpg” untuk mengerjakan soal pada nomor ini. Berikut
contoh keluarannya.

```python
# Load the image
image_path = '/content/drive/MyDrive/PCVK/Images/Face Detection/people.jpg'
img = cv.imread(image_path)
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Load the pre-trained face and eye cascade classifiers
face_cascade = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_eye.xml')

# Check if classifiers are loaded successfully
if face_cascade.empty() or eye_cascade.empty():
    print("Error loading cascade classifiers.")
    exit()

# Convert the image to grayscale for face and eye detection
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Detect faces in the image
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

# Print the number of detected faces
print(f"Number of faces detected: {len(faces)}")

# Loop through detected faces
for (x, y, w, h) in faces:
    cv.rectangle(img_rgb, (x, y), (x+w, y+h), (0, 0, 0), 2)  # Hitam untuk bounding box wajah

    # Extract the region of interest (ROI) for eye detection
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]

    # Adjust the eye detection parameters
    eyes = eye_cascade.detectMultiScale(roi_gray, scaleFactor=1.1, minNeighbors=3, minSize=(20, 20))

    # Print the number of detected eyes
    print(f"Number of eyes detected: {len(eyes)}")

    # Loop through detected eyes
    for (ex, ey, ew, eh) in eyes:
        cv.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (255, 255, 255), 2)  # Putih untuk bounding box mata

        # Detect right eye (assumption: the right eye is on the right side of the face)
        if ex > w/2:
            cv.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (255, 0, 0), 2)  # Merah untuk bounding box mata kanan

# Display the result
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
