```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
path = '/content/drive/MyDrive/PCVK/Images/kitten01.jpg'
img = cv.imread (path)
plt.imshow(img)
```

```python
# 2. Tampilkan image dalam channel Red-Blue dan Green-Blue saja!

# Pisahkan saluran warna
blue_channel = img[:, :, 0]  # Saluran biru
green_channel = img[:, :, 1]  # Saluran hijau
red_channel = img[:, :, 2]  # Saluran merah

# Buat gambar baru hanya dengan saluran warna RB (Red-Blue)
rb_image = img.copy()
rb_image[:, :, 1] = 0  # Nolkan saluran hijau (Green)

# Buat gambar baru hanya dengan saluran warna GB (Green-Blue)
gb_image = img.copy()
gb_image[:, :, 2] = 0  # Nolkan saluran merah (Red)

# Tampilkan gambar
plt.figure(figsize=(12, 4))
plt.subplot(132), plt.imshow(rb_image[..., ::-1]), plt.title('Red-Blue (RB)')
plt.subplot(133), plt.imshow(gb_image[..., ::-1]), plt.title('Green-Blue (GB)')
plt.show()
```

```python
# 3. Tampilkan image baris ke 20-115, kolom 25-120!

# Ambil bagian yang diinginkan dari gambar
cropped_img = img[20:116, 25:121]

# Tampilkan bagian yang diambil
plt.imshow(cropped_img[..., ::-1])  # Menampilkan dalam format RGB
plt.show()
```

```python
# 4. Tampilkan image baris ke 5-30, semua kolom, channel Red saja!

# Ambil saluran warna Merah (Red) dari bagian yang diinginkan
cropped_red_channel = img[5:31, :, 2]  # Saluran Merah (Red) adalah saluran ketiga (indeks 2)

# Tampilkan saluran warna Merah (Red)
plt.imshow(cropped_red_channel, cmap='gray')  # Tampilkan dalam skala warna abu-abu
plt.show()
```
